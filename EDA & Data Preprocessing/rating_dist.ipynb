```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
file_list = os.listdir(path)
model_dict = {file[:-4]: 
              pd.DataFrame({'rating': list(map(int, np.round(pd.read_csv(path + file)['rating'].values)))}) for file in file_list}
print(file_list)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()

for i, (model_name, rating) in enumerate(model_dict.items()):
    temp_df = model_dict[model_name].value_counts()
    temp_df = temp_df.reset_index()
    temp_df.columns = ['rating', 'count']
    print('--------------------------------------------')
    print(model_name)
    print(temp_df)
    
    axes[i].bar(temp_df['rating'], temp_df['count'])
    axes[i].set_title(model_name)

plt.show()
```
